```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("Spaceship_Titanic.csv")
data
```

```python
data.info(memory_usage="deep")
```

```python
data.isna().sum()
```

```python
most_common_planet=data["HomePlanet"].mode()[0]
most_common_planet
data["HomePlanet"].fillna(most_common_planet,inplace=True)
```

```python
most_common_crysleep=data["CryoSleep"].mode()[0]
data["CryoSleep"].fillna(most_common_crysleep,inplace=True)
```

```python
data.isna().sum()
```

```python
data.Cabin.apply(lambda x: (x.split("/") if isinstance(x,str) else [np.nan] * 3)[2]).value_counts()
```

```python
data.Cabin=data.Cabin.apply(lambda x: (x.split("/") if isinstance(x,str) else [np.nan] * 3)[2]=="S")
```

```python
most_common_destination=data.Destination.mode()[0] 
data.Destination.fillna(most_common_destination,inplace=True) 
```

```python
median_age=data["Age"].median()
data["Age"] = data["Age"].fillna(median_age).astype("int8")
```

```python
most_common_vip_status=data.VIP.mode()[0]
data["VIP"] = data["VIP"].fillna(most_common_vip_status)
```

```python
service=["RoomService","FoodCourt","ShoppingMall","Spa","VRDeck"]
data.loc[data.CryoSleep==True,service]=0

for col in service:
    median=data[col].median()
    data[col]=data[col].fillna(median)
```

```python
data.isna().sum()
```

```python
data=data.drop("Name",axis=1)
data
```

```python
data.isna().sum()
```

```python
data["PassengerId"].apply(lambda x:int(x.split("_")[1])).value_counts()
```

```python
data.PassengerId=data.PassengerId.apply(lambda x: int(x.split("_")[1]=="01"))
```
