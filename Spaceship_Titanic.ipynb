```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
```

```python
data=pd.read_csv("Spaceship_Titanic.csv")
data
```

```python
data.info(memory_usage="deep")
```

```python
data.isna().sum()
```

```python
most_common_planet=data["HomePlanet"].mode()[0]
most_common_planet
data["HomePlanet"].fillna(most_common_planet,inplace=True)
```

```python
most_common_crysleep=data["CryoSleep"].mode()[0]
data["CryoSleep"].fillna(most_common_crysleep,inplace=True)
```

```python
data.isna().sum()
```

```python
data.Cabin.apply(lambda x: (x.split("/") if isinstance(x,str) else [np.nan] * 3)[2]).value_counts()
```

```python
data.Cabin=data.Cabin.apply(lambda x: (x.split("/") if isinstance(x,str) else [np.nan] * 3)[2]=="S").astype("int64")
```

```python
most_common_destination=data.Destination.mode()[0] 
data.Destination.fillna(most_common_destination,inplace=True) 
```

```python
median_age=data["Age"].median()
data["Age"] = data["Age"].fillna(median_age).astype("int8")
```

```python
most_common_vip_status=data.VIP.mode()[0]
data["VIP"] = data["VIP"].fillna(most_common_vip_status).astype("int64")
```

```python
service=["RoomService","FoodCourt","ShoppingMall","Spa","VRDeck"]
data.loc[data.CryoSleep==True,service]=0

for col in service:
    median=data[col].median()
    data[col]=data[col].fillna(median)
data["CryoSleep"]=data["CryoSleep"].astype("int64")    
```

```python
data.isna().sum()
```

```python
data=data.drop("Name",axis=1)
data
```

```python
data.isna().sum()
```

```python
data["PassengerId"].apply(lambda x:int(x.split("_")[1])).value_counts()
```

```python
data.PassengerId=data.PassengerId.apply(lambda x: int(x.split("_")[1]=="01"))
```

```python
data
```

```python
data.info(memory_usage="deep")
```

```python
data = pd.get_dummies(data,columns=["HomePlanet"],drop_first=False)
planets=["HomePlanet_Earth","HomePlanet_Europa","HomePlanet_Mars"]
data[planets]=data[planets].astype(int)
```

```python
data = pd.get_dummies(data,columns=["Destination"])
dummy_cols = [c for c in data.columns if c.startswith("Destination_")]
data[dummy_cols]=data[dummy_cols].astype(int)
```

```python
data
```

```python
y=data["Transported"]
x=data.drop(columns="Transported")
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
model=LogisticRegression()
model.fit(x_train,y_train)
y_pred_test=model.predict(x_test)
submission = pd.DataFrame({
    "PassengerId": x_test["PassengerId"],  
    "Transported": y_pred_test
})

submission.to_csv("submission.csv", index=False)

```

```python
accuracy=accuracy_score(y_test,y_pred_test)
accuracy
```

```python
model=KNeighborsClassifier(n_neighbors=15)
model.fit(x_train,y_train)
```

```python
y_pred_test=model.predict(x_test)
accuracy=accuracy_score(y_test,y_pred_test)
accuracy
```
