```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
data=pd.read_csv("Mall_Customers.csv")
```

```python
data
```

```python
data.info()
```

```python
data.drop(columns="CustomerID",axis=1,inplace=True)
```

```python
data["Gender"]=(data["Gender"]=="Male").astype("int8")
```

```python
data
```

```python
scaler=MinMaxScaler()
data_sc = pd.DataFrame(scaler.fit_transform(data),columns=data.columns,index=data.index)
data_sc
```

```python
model=KMeans(n_clusters=5,random_state=0)
clusters=model.fit_predict(data_sc)
data_sc["Cluster"]=clusters
data_sc
```

```python
score=silhouette_score(data_sc, labels=model.labels_, random_state=0)
score
```
