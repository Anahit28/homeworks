```python
import pandas as pd
import numpy as np

from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.neighbors import KNeighborsRegressor
```

```python
data=pd.read_csv("Student_Performance.csv")
data
```

```python
data.info(memory_usage="deep")
```

```python
y=data["Performance Index"]
x=data.drop("Performance Index",axis=1)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size=0.25, random_state=0)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
data["Hours Studied"]=data["Hours Studied"].astype("int8")
```

```python
data.info(memory_usage="deep")
```

```python
scaler=MinMaxScaler()
x_hs_train_sc=pd.DataFrame(scaler.fit_transform(x_train[["Hours Studied"]]),columns=["Hours Studied"],
    index=x_train.index)
x_hs_test_sc=pd.DataFrame(scaler.transform(x_test[["Hours Studied"]]),columns=["Hours Studied"],
    index=x_test.index)


x_train["Hours Studied"]=x_hs_train_sc
x_test["Hours Studied"]=x_hs_test_sc
```

```python
scaler=MinMaxScaler()
x_ps_train_sc=pd.DataFrame(scaler.fit_transform(x_train[["Previous Scores"]]),columns=["Previous Scores"],
    index=x_train.index)
x_ps_test_sc=pd.DataFrame(scaler.transform(x_test[["Previous Scores"]]),columns=["Previous Scores"],
    index=x_test.index)


x_train["Previous Scores"]=x_ps_train_sc
x_test["Previous Scores"]=x_ps_test_sc
x_train
```

```python
data.drop("Extracurricular Activities",axis=1,inplace=True)
```

```python
scaler=MinMaxScaler()
x_sleep_train_sc=pd.DataFrame(scaler.fit_transform(x_train[["Sleep Hours"]]),columns=["Sleep Hours"],index=x_train.index)
x_sleep_test_sc=pd.DataFrame(scaler.transform(x_test[["Sleep Hours"]]),columns=["Sleep Hours"],index=x_test.index)

x_train["Sleep Hours"]=x_sleep_train_sc
x_test["Sleep Hours"]=x_sleep_test_sc
```

```python
x_train
```

```python
d=x_train["Extracurricular Activities"].value_counts()
d
```

```python
x_train["Extracurricular Activities"] = (x_train["Extracurricular Activities"] == "Yes").astype("int8")
x_test["Extracurricular Activities"] = (x_test["Extracurricular Activities"] == "Yes").astype("int8")
```

```python
x_test
```

```python
scaler=MinMaxScaler()
x_pp_train_sc=pd.DataFrame(scaler.fit_transform(x_train[["Sample Question Papers Practiced"]]),columns=["Sample Question Papers Practiced"],index=x_train.index)
x_pp_test_sc=pd.DataFrame(scaler.transform(x_test[["Sample Question Papers Practiced"]]),columns=["Sample Question Papers Practiced"],index=x_test.index)

x_train["Sample Question Papers Practiced"]=x_pp_train_sc
x_test["Sample Question Papers Practiced"]=x_pp_test_sc
```

```python
x_train
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred_test=model.predict(x_test)
y_pred_test
```

```python
y_test.values
```

```python
mae=mean_absolute_error(y_test,y_pred_test)
mae
```
